```python
import os
import json
import shutil

# === DEBUG : Afficher le chemin courant ===
print("Current Working Directory:", os.getcwd())

# === Chemins ===
# Chemin vers l'ancien modèle (meilleur précédent)
old_config_path = r"C:\Users\HP\Desktop\sequencedemodel\modelsfinals\config.json.txt"

# Chemin vers le résultat temporaire du nouveau modèle
temp_result_path = "temp_results.json"

# Dossier source du meilleur checkpoint nouvellement entraîné
checkpoint_source = "checkpoints/checkpoint-best"

# Dossier destination pour le meilleur modèle
final_model_dir = r"C:\Users\HP\Desktop\sequencedemodel\modelsfinals"

# === Étape 1 : Charger l'ancien F1 ===
with open(old_config_path, "r", encoding="utf-8") as f:
    old_model_info = json.load(f)

old_f1 = old_model_info["f1_score"]
print(f"Ancien F1 : {old_f1}")

# === Étape 2 : Charger le nouveau F1 ===
with open(temp_result_path, "r", encoding="utf-8") as f:
    new_results = json.load(f)

new_f1 = new_results["eval_f1"]
print(f"Nouveau F1 : {new_f1}")

# === Étape 3 : Comparaison et sauvegarde si amélioration ===
if new_f1 > old_f1:
    print("✨ Nouveau modèle meilleur, copie en cours...")
    shutil.copytree(checkpoint_source, final_model_dir, dirs_exist_ok=True)
    print("✅ Modèle copié avec succès !")
else:
    print("🔁 Ancien modèle conservé.")
```
